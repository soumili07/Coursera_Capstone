```python
url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
page = urllib.request.urlopen(url)
```

```python
soup = BeautifulSoup(page, "lxml")
```

```python
print(soup.prettify())
```

```python
post_table = soup.find('table', class_='wikitable sortable')
post_table
```

```python
A=[]
B=[]
C=[]

for row in post_table.findAll('tr'):
    cells=row.findAll('td')
    if len(cells) == 3:
        A.append(cells[0].find(text=True).strip())
        B.append(cells[1].find(text=True).strip())
        C.append(cells[2].find(text=True).strip())
```

```python
import pandas as pd
df=pd.DataFrame(A,columns=['PostalCode'])
df['Borough']=B
df['Neighborhood']=C
df.head()
```

```python
df2 = df1
df2.head()
df2.shape
df2 = pd.DataFrame(df2)
df2.shape
```

# csv file to create the following new dataframe with latitude and longitude:

```python
# define the dataframe columns
column_names = ['PostalCode', 'Borough', 'Neighborhood', 'Latitude', 'Longitude'] 

# instantiate the dataframe
new_dataset = pd.DataFrame(columns=column_names)
for data in df2.iterrows():
    #print(data[1]['PostalCode'])
    #print(df_position[(df_position.PostalCode == data[1]['PostalCode'])]['Latitude'].values[0])
    new_dataset = new_dataset.append({'PostalCode': data[1]['PostalCode'],
                                      'Borough': data[1]['Borough'],
                                      'Neighborhood': data[1]['Neighborhood'],
                                      'Latitude': df_position[(df_position.PostalCode == data[1]['PostalCode'])]['Latitude'].values[0],
                                      'Longitude': df_position[(df_position.PostalCode == data[1]['PostalCode'])]['Longitude'].values[0]
                                     }, ignore_index=True)
    
new_dataset.head()
    
```

```python
address = 'Toronto, Ontario'

geolocator = Nominatim(user_agent="ny_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronto, Ontario are {}, {}.'.format(latitude, longitude))
```

# Data Visualization

```python
# create map of New York using latitude and longitude values
map_toranto = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for lat, lng, borough, neighborhood in zip(new_dataset['Latitude'], new_dataset['Longitude'], new_dataset['Borough'], new_dataset['Neighborhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toranto)  
    
map_toranto
```
